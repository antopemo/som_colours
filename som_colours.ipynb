```python
from GEMA import Map, Visualization
import numpy as np
from matplotlib import pyplot as plt
from sklearn import preprocessing
```

First we define a set of 250 different colors for the map to learn.

```python
colors = np.random.randint(0, 255, (250, 3))
colors = preprocessing.normalize(colors)
```

We create a Map Object and train it. At first the map would be chaotic, but as it trains it starts ordering the colours.

```python
som = Map(data=colors, 
          size= 10, 
          period = 500)
```

Once we have created and trained the map, we can inspect the weight matrix, represented as RGB colors.

```python
plt.imshow(som.weights)
```

Now we can inspect by each colour channel, so we get an idea of how each vector component is layed out on the map. This concept is known as Codebook vector in the SOM ethimology, and it can be very useful to check if the map trained well or to see where a value for a component should be located in the map.

```python
red = som.weights.copy()
red[:,:,1], red[:,:,2] = 0, 0
plt.imshow(red)
```

```python
green = som.weights.copy()
green[:,:,0], green[:,:,2] = 0, 0
plt.imshow(green)
```

```python
blue = som.weights.copy()
blue[:,:,0], blue[:,:,1] = 0, 0
plt.imshow(blue)
```
